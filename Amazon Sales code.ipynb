```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
amazon_sales.rename(columns={"Item Type":"Item_Type","Sales Channel":"Sales_Channel","Order Priority":"Order_Priority","Order Date":"Order_Date","Order ID":"Order_ID","Ship Date":"Ship_Date", 'Units Sold':'Units_Sold', 'Unit Price':'Unit_Price','Unit Cost':'Unit_Cost', 'Total Revenue':'Total_Revenue', 'Total Cost':'Total_Cost', 'Total Profit':'Total_Profit'},inplace=True)
```

```python
amazon_sales.columns
```

```python
plt.figure(figsize=(17,13),dpi= 1000)
profit_by_region = amazon_sales.groupby('Region')['Total_Profit'].sum().reset_index()
plt.bar(profit_by_region['Region'], profit_by_region['Total_Profit'], color='brown')
plt.xlabel('Region')
plt.ylabel('Total Profit')
plt.title('Total Profit by Region')
plt.xticks(rotation=45)
plt.rcParams.update({'font.size': 30})
plt.savefig('bar_chart_example.png', dpi=1000)
plt.tight_layout()
plt.show()
```

```python
sales_channel_counts = amazon_sales['Sales_Channel'].value_counts()

plt.figure(figsize=(6, 6))
plt.pie(sales_channel_counts, labels=sales_channel_counts.index, autopct='%1.1f%%', startangle=140,colors='brown')
plt.title('Sales Channel Distribution')
plt.axis('equal')
plt.savefig('pie_chart_example.png', dpi=1000)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(amazon_sales['Unit_Price'], amazon_sales['Total_Profit'], color='brown', alpha=0.6)
plt.xlabel('Unit Price')
plt.rcParams.update({'font.size': 7})
plt.ylabel('Total Profit')
plt.title('Profit vs Unit Price')
plt.savefig('scatter_chart_example.png', dpi=1000)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.hist(amazon_sales['Total_Cost'], bins=10, color='chocolate', edgecolor='black')
plt.xlabel('Total Cost')
plt.ylabel('Frequency')
plt.title('Distribution of Total Cost')
plt.savefig('hist_chart_example.png', dpi=1000)
plt.tight_layout()
plt.show()
```
